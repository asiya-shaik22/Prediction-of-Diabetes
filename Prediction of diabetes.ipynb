```python
# C. Check Proportion for diabetes vs non diabetes
num_true = len(df.loc[df['diabetes'] == True])
num_false = len(df.loc[df['diabetes'] == False])
print(f'num_true = {num_true}')
print(f'num_false = {num_false}')

per_num_true = (num_true/(num_true+num_false))*100
print(per_num_true)

per_num_false = (num_false/(num_true+num_false))*100
print(per_num_false)

print('True = ',(num_true/(num_true+num_false) )*100)
print('False = ',(num_false/ (num_true+num_false) )*100 )
```

```python
#Step 3: Train test split
```

```python
from sklearn.model_selection import train_test_split
input_columns = ['num_preg', 'glucose_conc', 'diastolic_bp', 'insulin',
                   'bmi', 'diab_pred', 'age',  'skin']

output_column =['diabetes']
x = df[input_columns]
y = df[output_column]
split_test_size = 0.3
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=split_test_size,
                                                 random_state=42)
```

```python
print("X_train",x_train)
print('x_test',x_test)
print('y_train',y_train)
print('y_test',y_test)
```

```python
# Check the shape of splitted data
print("X_train",x_train.shape)
print('x_test',x_test.shape)
print('y_train',y_train.shape)
print('y_test',y_test.shape)
```

```python
# Check the proportion of training and testing after splitting
print('{0:0.2f}% in training set'.format((len(x_train)/len(df.index))*100))
print('{0:0.2f}% in training set'.format((len(x_test)/len(df.index))*100))
```

```python
# Check the diabetes and non diabetes cases in training and testing
print('training true:{0} ({1:0.2f}%)'.format(len(y_train[y_train[ : ] == 1]),
                                              len(y_train[y_train[ : ] ==1])/len(y_train) * 100))
print('training False:{0} ({1:0.2f}%)'.format(len(y_train[y_train[ : ] == 0]),
                                              len(y_train[y_train[ : ] ==0])/len(y_train) * 100))
print('Test true:{0} ({1:0.2f}%)'.format(len(y_test[y_test[ : ] == 1]),
                                         len(y_test[y_test[ : ] ==1])/len(y_train) * 100))
print('Test false:{0} ({1:0.2f}%)'.format(len(y_test[y_test[ : ] == 0]),
                                          len(y_test[y_test[ : ] ==0])/len(y_train) * 100))
```

```python
from sklearn.impute import SimpleImputer
fill_values = SimpleImputer(missing_values=0,strategy='mean')
x_train = fill_values.fit_transform(x_train) # fit_transform method fills all the zeros with mean values
x_test = fill_values.fit_transform(x_test)
```

```python
x_train[0:5]
```

```python
# Step 5: Training the algorithms
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model = GaussianNB() # creating object for GaussianNB class
nb_model.fit(x_train,y_train.values.ravel()) # Training a model
```

```python
from sklearn import metrics
nb_predict_model = nb_model.predict(x_train) # Testing a model
print(f'Predicted accuracy { metrics.accuracy_score(y_train,nb_predict_model) }')
```

```python
from sklearn import metrics
nb_predict_test = nb_model.predict(x_test)
print(f'Predicted accuracy { metrics.accuracy_score(y_test,nb_predict_test)}')
```

```python
print(f'{metrics.confusion_matrix(y_test,nb_predict_test)}')
```

```python
print(f'{ metrics.classification_report(y_test,nb_predict_test)}')
```

```python
from sklearn.ensemble import RandomForestClassifier
rf_model = RandomForestClassifier(random_state=42) # Creating object to RandomforestClassifier class
rf_model.fit(x_train,y_train.values.ravel()) # Training a machine

```

```python
from sklearn import metrics
rf_predict_test= rf_model.predict(x_test) # Testing a model
print(f'Predicted accuracy { metrics.accuracy_score(y_test,rf_predict_test) }')
```

```python
from sklearn import metrics
rf_predict_model = rf_model.predict(x_train) # Testing a model
print(f'Predicted accuracy { metrics.accuracy_score(y_train,rf_predict_model) }')
```

```python
print(f'{metrics.confusion_matrix(y_test,rf_predict_test)}')
```

```python
print(f'{metrics.classification_report(y_test,rf_predict_test)}')
```

```python
from sklearn.neighbors import KNeighborsClassifier

k_neighbor = KNeighborsClassifier()
k_neighbor.fit(x_train,y_train.values.ravel())
```

```python
x_pred = k_neighbor.predict(x_test)

print(f'Accuracy = {metrics.accuracy_score(y_test,x_pred)}')
print(f'Confusion matrix = \n{metrics.confusion_matrix(y_test,x_pred)}')
```

```python
print(f'Classification report =\n{metrics.classification_report(y_test,x_pred)}')
```
